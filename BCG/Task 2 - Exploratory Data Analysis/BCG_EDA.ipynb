## Group price_df by the mean for each company id and merge client_df to it

```python
agg_df=price_df.groupby('id').mean()
```

```python
df=pd.merge(client_df, agg_df, on='id', how='left')
```

## Data Visualization

```python
plt.title('Churn Distribution')
plt.pie(df.churn.value_counts(),labels=['Retained','Churned'], autopct='%1.1f%%',explode=[0,.05]);
```

```python
consumption=[i for i in df.columns if 'cons' in i]+['churn']
```

```python
consumption
```

```python
row = int((len(consumption)-1)/2+(len(consumption)-1)%2)
fig, axes = plt.subplots(row, 2, figsize=(14, 10), layout='tight')
plt.suptitle('Enegy and Forecasted Energy Consumption Distribution')
for num, col in enumerate(consumption[:-1]):
    sns.histplot( x=df[col], hue= df['churn'], bins=20, multiple='stack', ax=axes[num//2][num%2])
    axes[num//2][num%2].legend(title='Churn Status',labels=['Churned','Retained'])
    axes[num//2][num%2].set_title(col+' Distribution')
```

```python
price=[i for i in df.columns if 'price' in i]+['churn']
```

```python
row = int((len(price)-1)/2+(len(price)-1)%2)
fig, axes = plt.subplots(row, 2, figsize=(14, 12), layout='tight')
plt.suptitle('Price and Forecasted Price Distribution')
for num, col in enumerate(price[:-1]):
    sns.histplot( x=df[col], hue= df['churn'], bins=20, multiple='stack', ax=axes[num//2][num%2])
    axes[num//2][num%2].legend(title='Churn Status',labels=['Churned','Retained'])
    axes[num//2][num%2].set_title(col+' Distribution')
```

```python
categories=df.select_dtypes('object').columns.drop('id')
categories=list(categories)+['churn']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
plt.suptitle('Churn per Category')
for num, col in enumerate(categories[:-1]):
    sns.countplot(data=df, x=col, hue='churn', ax=axes[num])
    axes[num].legend(title='Churn Status',labels=['Retained','Churnedb'])
    axes[num].set_title(col+' Churn')
    axes[num].set_xticklabels(axes[num].get_xticklabels(),rotation=90)
```

```python
others=df[list(set(df.columns)-set(categories+price+consumption))].select_dtypes(['int','float']).columns
others=list(others)+['churn']
```

```python
row = int((len(others)-1)/2+(len(others)-1)%2)
fig, axes = plt.subplots(row, 2, figsize=(14, 12), layout='tight')
plt.suptitle('Distribution of other Variables')
for num, col in enumerate(others[:-1]):
    sns.histplot( x=df[col], hue= df['churn'], bins=20, multiple='stack', ax=axes[num//2][num%2])
    axes[num//2][num%2].legend(title='Churn Status',labels=['Churned','Retained'])
    axes[num//2][num%2].set_title(col+' Distribution')
```
